## 1. Podstawowa analiza rynku kryptowalut
<p>Głównym celem będzie wskazanie typów cryptowalut najbardziej przydatnych do inwestowania. W domyśle są to kryptowaluty o małej wartości rynkowej. Pośrednio pokażemy takie elementy, jak duży jest rynek kryptowalut, jak BTC wygląda na tle pozstałych kryptowalut, jak często zmieniaja się wartości walut wirtualnych, które kryptowaluty umożliwiają najwyższy zysk w badanym okresie czasu. Wprowadzimy też kilka elementów które będą oparte na teorii wizualizacji danych.</p>
<p>Projekt jest wykonałem na podstawie szkolenia dostępnego na portalu <a href="https://datacamp.com">datacamp</a>
<p>W analizie wykorzystany jest plik CSV z API coinmarketcap z 6 grudnia 2017 roku.</p>

```python
# import pandasa
import pandas as pd

# import matplotlibwraz z ustawieniami do renderu wykresow
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg' 
plt.style.use('fivethirtyeight')

# wczytanie csvki do pandasa
dec6 = pd.read_csv('datasets/coinmarketcap_06122017.csv')

# sprawdzenie ilosci WSZYSTKICH danych na podstawie 2 kolumn
market_cap_raw = dec6[['id','market_cap_usd']]
print(market_cap_raw.count())
```

## 2. Odfiltrowanie kryptowalut bez wartości rynkowej
<p>Powyższy wynik pokazuje że mamy kryptowaluty bez wartości rynkowej (kapitalizacji). W analizie wartości opartej na dolarze, należy je wykluczyć</p>

```python
# Filter krytpowalut bez wartości rynkowej
cap = market_cap_raw.query('market_cap_usd > 0')

# Ponowne sprawdzenie ilości danych
print(cap.count())
```

## 3. Wartość rynkowa BTC vs pozostałe kryptowaluty
<p>Najpopularniejszą kryptowalutą jest obecnie Bitcoin (BTC). Pomimo towrzenia nowych kryptowalut, jego pozycja jako głównego środka płatniczego i do kapitalizacji. Poniżej wizualizacja pokazująca lepiej wartość BTC na tle pozostałych kryptowalut</p>

```python
#Deklaracj tytułu i nazwy osi
TOP_CAP_TITLE = 'Top 10 krytowlaut wg. wartości rynkowej'
TOP_CAP_YLABEL = '% całości'

# Pozostawienie 10 najpopularniejszych kryptowalut
cap10 = cap.set_index('id')
cap10 = cap10.iloc[0:10]

# Obliczanie wartości procentowej
cap10 = cap10.assign(market_cap_perc = lambda x: (x.market_cap_usd / cap.market_cap_usd.sum()) * 100 ) 

# Podpinanie tytułu, oraz nazwy osi Y
ax = cap10['market_cap_perc'].plot(kind='bar', title=TOP_CAP_TITLE)
ax.set_ylabel(TOP_CAP_YLABEL)
ax.set_xlabel("")
```

Powyższy wykres pokazuje, że rynki kryptowalut, są zdominowane przez Bitcoina i Etherum.

## 4. Poprawa wizualizacji wykresu
<p>Powyższy wykres nie spełnia podstawowych założeń wizualizacji. Należy poprawić nast. elementy:</p>
<ul>
    <li>Skala wykresu jest nieczytelna, BTC zdominował wykres i skala dla pozostałych kryptowalut przestała być czytelna</li>
    <li>Jednolita kolorystyka utrudnia łatwe rozróznianie kryptoalut</li>
</ul>

```python
# Kolory, kryptowaluty oparte o BTC zostały zaznaczone jednym kolorem
COLORS = ['orange', 'green', 'orange', 'cyan', 'yellow', 'blue', 'silver', 'orange', 'red', 'purple']

# Ponowne generowanie wykresu wraz z kolorami i skalowaniem logarytmicznym
ax = cap10.market_cap_usd.plot(kind='bar', color=COLORS)
ax.set_yscale('log')

# Poprawa tytułu, oraz osi X i Y
ax.set_title('Top 10 rynków krytowlaut wg. wartości rynkowej')
ax.set_ylabel("USD")
ax.set_xlabel('')
```

Nie licząc rozwiązań opartych na BTC, na rynku oprócz Etherum liczą się min waluty takie jak Ripple, Dash, Litecoin czy Monero. Rynek Cardano ma już niewielką wartość rynkową, o jest dużo lepiej widoczne, niż na wykresie z punktu 3.

## 5. Dynamika i stopień zmienności 
<p>Jednym z największym wyrózników rynku kryptowalut na tle innych rynków inwestycyjnych, jest jego dynamika. Dzięki temu możliwe jest uzyskiwanie spektakularnych zysków w krótkim okresie czasu.</p>
<p>Sprawdzamy dynamikę i wyróżniamy kryptoawaluty o największym stopniu zmienności w okresie dniowym i tygodniowym.</p>

```python
# Wybieramy interesujące elementy z dataframe, w tym przypadku ID, oraz zmiany dzienne i tygodniowe
volatility = dec6[['id','percent_change_24h','percent_change_7d']]

# Spolszczamy tytuły kolumn
volatility.columns = ['id','dynamika_dzienna','dynamika_tygodniowa']

# Ustawiamy index dataframe na id, oraz usuwamy kryptowaluty bez określonej zmienności. Sortujemy wg. zmiany dziennej
volatility = volatility.set_index('id')
volatility = volatility.dropna(0, how='any')
volatility = volatility.sort_values('dynamika_dzienna')

# Zwracamy topowe i końcowe wartości
print(volatility.head(10))
print(volatility.tail(10))
```

## 6. Dzienna dynamika wartości
<p>Powyższe dane nie są miarodajne, ponieważ nie uwzględniają realnej wartości rynków. Kryptowaluty o małej wartości rynkowej charakteryzują się większą dynamiką. W celu pokazania realnych zysków i strat dla danej kryptowaluty, musimy zestawić ww. wartości z realną wartością danego rynku kryptowalut.</p>

```python
#Definicja funkcji (użyjemy jej 2 razy, najpierw do dziennej, potem tygodniowej)
def top10_subplot(volatility_series, title):
    # Deklarowanie 2 wykresów
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))
    
    # Wykres 10 największych strat
    ax = volatility_series[:10].plot.bar(color="darkred", ax=axes[0])
    
    # Nadawanie tytułu (jest deklarowany poza funkcją, aby móc użyć funkcji ponownie)
    fig.suptitle(title)
    
    # Zmiana tytułu dla osi Y
    ax.set_ylabel('% change')
    
    # 10 największych zysków
    ax = volatility_series[-10:].plot.bar(color="darkblue", ax=axes[1])
    
    return fig, ax

DTITLE = "Największe wzrosty i straty rynków kryptowalut w ujęciu dziennym"

# CWywołanie funkcji z nazwą wykresu 
fig, ax = top10_subplot(volatility.dynamika_dzienna, DTITLE)
```

## 7. Tygodniowa dynamika wartości 
<p>Wykresy w punkcie 6, oparte są na danych dziennych, co drastycznie wpłynęło na wartości. Powyższe dane nie mają szczególnej wartości, przy wyborze kryptowalut w celach inwestycyjnych długofalowych, czy nawet osiągania zysków z shortów.</p>
<p>W celu uzyskania bardziej użytecznych danych, skorzystamy z wykresów opartych na danych tygodniowych</p>

```python
# Sortowanie danych wg dynamiki tygodniowej
volatility7d = volatility.sort_values("dynamika_tygodniowa")

WTITLE = "Największe wzrosty i straty rynków kryptowalut w ujęciu tygodniowym"

# Wywoływanie funkcji z punktu 6 z nowymi danymi i tytułem
fig, ax = top10_subplot(volatility7d.dynamika_tygodniowa, WTITLE);
```

Na podstawie powyższych danych, możemy już łatwiej określić które z kryptowalut mogą być interesujące jeśli chodzi o inwestowanie w krótkim okresie czasu. Wykluczyliśmy dzięki temu, min. FedoraCoin, które w poprzednim punkcie było pokazane jako kryptowaluta o największym wzroście wartości rynku.

## 8. Kryptowaluty na tle "prawdziwej" giełdy
<p>Praktycznie wszystkie kryptowaluty z punktu 6 i 7 są mało znane i nie wystąpiły na wykresie z pkt. 4. To tylko udowadnia tezę która była tłem analizy, czyli jakoby to mało znane kryptowaluty były najlepszym sposobem na inwestycję, a nie popularny Bitcoin</p>
<p>Popularny serwis dotyczący inwestycji giełdowych, określa sposób kategoryzacji akcji spółek występujących na giełdzie. Porównamy dane dotyczące rynków kryptowalut do ww. definicji</p>
<p>Poniżej tzw. <a href="https://www.investopedia.com/terms/l/large-cap.asp">Large Cap</a>, w przypadku tradycyjnej giełdy byłyby to akcje spółek o wartości ponad 10 miliardów dolarów</p>

```python
# Filtr i zwrócenie wszystkiego o wartości ponad 10 mld dolarów
largecaps = cap.query("market_cap_usd > 1E+10")
largecaps
```

Jak widać, w definicji Largecap (na realnej giełdzie należątu np. Facebook, Amazon, czy Apple), mieszczą się tylko 4 waluty. Poniżej podział rynków krytpowalut 3 kategorie, w zależności od wartości. Zastosowany został inny przedział, aby zwiększyć czytelnośc.

```python
def capcount(query_string):
    return cap.query(query_string).count().id

# Labels for the plot
LABELS = ["Największe", "Średnie", "Małe"]

# Using capcount count the not_so_small cryptos
biggish = capcount("market_cap_usd > 3E+8")

# Same as above for micro ...
micro = capcount("market_cap_usd >= 5E+7 & market_cap_usd < 3E+8")

# ... and for nano
nano =  capcount("market_cap_usd < 5E+7")

# Making a list with the 3 counts
values = [biggish, micro, nano]

# Plotting them with matplotlib 
plt.bar(range(len(values)), values, tick_label=LABELS);
```
